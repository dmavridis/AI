## A* Search
Combines the advantages of Breadth first and Greedy best-first search 

$$f = g + h$$
g(path) = path cost

h(path) = h(s) = Estimated distance to the goal

Search strategy that is the best possible. Finds the shortest length path while expanding minimum 

A* finds the lower cost path if:
- It depends on the h function $ h(s) < true cost$
- h should never overestimate distance to goal
- h is optimistic
- h is admissible
